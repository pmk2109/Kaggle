```python
from bayes_opt import BayesianOptimization
from sklearn.cross_validation import cross_val_score

def xgbclassifiercv(max_depth, learning_rate, n_estimators, gamma, min_child_weight, max_delta_step, subsample,
              colsample_bytree, silent =True, nthread = -1, seed = 1234):
    
    return cross_val_score(XGBClassifier(max_depth = int(max_depth), learning_rate = learning_rate,
                                         n_estimators = int(n_estimators), silent = silent,
                                         nthread = nthread, gamma = gamma, min_child_weight = min_child_weight,
                                         max_delta_step = max_delta_step, subsample = subsample,
                                         colsample_bytree = colsample_bytree, seed = seed,
                                         objective = 'reg:logistic'),
                           X_train,
                           y_train,
                           'log_loss',
                           n_jobs=-1,
                           cv=3).mean()

xgbclassifierBO = BayesianOptimization(xgbclassifiercv,
                                 {'max_depth': (5, 12),
                                  'learning_rate': (0.001, 0.5),
                                  'n_estimators': (50, 1000),
                                  'gamma': (1., 0.01),
                                  'min_child_weight': (1, 100),
                                  'max_delta_step': (0, 10),
                                  'subsample': (0.01, 0.9),
                                  'colsample_bytree' :(0.5, 0.99)
                                 })
xgbclassifierBO.maximize()
print('-'*53)
print('Final Results')
print('XGBlinear: %f' % xgbclassifierBO.res['max']['max_val'])
```

```python
import pandas as pd
import numpy as np
DATA_DIR = "/Users/patrickkennedy/Desktop/Data_Science_Training/Kaggle"
train = pd.read_csv(DATA_DIR + "/BNP_Paribas/train.csv")
test = pd.read_csv(DATA_DIR + "/BNP_Paribas/test.csv")
categorical = []

for col in train.columns:
    s = str(train[col][0])
    d = str(train[col][1])
    f = str(train[col][2])
    try:
        float(s)
        float(d)
        float(f)
    except ValueError:
        categorical.append(col)
for cat in categorical:
    train[cat] = pd.factorize(train[cat])[0]


train.fillna(-1, inplace=True)

X = train.drop(["ID", "target"], axis=1)
y = train.target



```

```python
columns = list(X.columns)
for i, col in enumerate(columns):
    for j in columns[i:]:
        X[col+j]=X[col]*X[j]
        #print i, col, j
    
    
        
```

```python
from sklearn.feature_selection import SelectKBest, SelectFpr, chi2, f_classif
from sklearn.cross_validation import train_test_split

sample_weight = np.random.RandomState(42).rand(train["target"].shape[0])

#split train and test for calibration
X_train, X_test, y_train, y_test, sw_train, sw_test = \
    train_test_split(train.drop(["ID", "target"], axis=1), 
                     train["target"], 
                     sample_weight, 
                     test_size=0.33, 
                     random_state=42)

    
#one approach may be to jitter the random state and take 10 results of one model and avg them together    




import xgboost as xgb


params = { 
   "objective": "binary:logistic",
   #"booster": "gbtree",
   "eval_metric": "logloss",
   "eta": 0.01, # 0.06,
   #"min_child_weight": 240,
   "subsample": 0.75,
   "colsample_bytree": 0.68,
   "max_depth": 7
}
```

```python
for k in [4, 8, 12, 15, 20, 25, 30, 40, 50, 60, 75, 100, 125, 150, 175, 200]:
    
    X_new = SelectKBest(f_classif, k).fit(X, y).transform(X)
    X_train, X_test, y_train, y_test, sw_train, sw_test = \
        train_test_split(X_new, y, sample_weight, test_size=0.33, random_state=42)
    #how do i know which variables are which for my test set?

    xgtrain = xgb.DMatrix(X_train, y_train)
    xgtest = xgb.DMatrix(X_test)

    boost_round = 1000
    clf = xgb.train(params, xgtrain, num_boost_round=boost_round, verbose_eval=True, maximize=False)
    pred6 = clf.predict(xgtest, ntree_limit=clf.best_iteration)
    print log_loss(y_test, pred6), k
```

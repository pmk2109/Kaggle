```python
from sklearn.cross_validation import train_test_split

train.fillna(-1, inplace=True)

categorical = ['v3', 'v22', 'v24', 'v30', 'v31', 'v38', 'v47', 'v52', 'v56', 'v62', 'v66', 'v71', 'v74', 'v75', 'v79', 'v91', 'v107', 'v110', 'v112', 'v113', 'v125']
#print train[categorical] #= pd.factorize(train[categorical])[0]

for cat in categorical:
    train[cat] = pd.factorize(train[cat])[0]

X_train, X_test, y_train, y_test = train_test_split(train.drop(["ID", "target"], axis=1), train.target, test_size=0.2, random_state=42)
X_train = X_train.as_matrix()
y_train = y_train.values

#pd.set_option("display.max_columns", None)


#train.drop(['ID', 'target'], axis=1)
```
